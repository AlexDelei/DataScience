```python
import pandas
minfo = pandas.read_csv('ratings.csv')
minfo.head(10)
```

```python
mtitle = pandas.read_csv('movietitles.csv')
mtitle.head(10)
```

```python
data = pandas.merge(minfo , mtitle , on = 'item_id')
data.head(10)
```

```python
data.groupby('title')['rating'].mean().sort_values(ascending=False)
```

```python
data.groupby('title')['rating'].count().sort_values(ascending = False)
```

```python
ratings = pandas.DataFrame(data.groupby('title')['rating'].mean())
ratings['num_of_ratings'] = data.groupby('title')['rating'].count()
ratings.head(100)
```

```python
pivot = data.pivot_table(index= 'user_Id', columns = 'title', values = 'rating')
pivot.head(10)
```

```python
selected_movie = pivot['Godfather: Part II, The (1974)']
selected_movie.tail(15)
```

```python
similar = pivot.corrwith(selected_movie)
similar_df = pandas.DataFrame(similar, columns = ['correlation'])
similar_df.sort_values('correlation', ascending = False).head(15)
```

```python
similar_df = similar_df.join(ratings ['num_of_ratings'])
similar_df[similar_df['num_of_ratings'] > 100].sort_values('correlation', ascending = False).head(20)
```
